```python
muertes = pd.read_excel('MUERTES-CANMAMA-M40-MUN-COL.xlsx')
muertes
```

```python
muertes_comb = muertes.groupby(['Depto', 'MpNombre']).ngroups
muertes_comb
```

```python
# Combinaciones únicas en pob
pob_combinaciones = set(pob[['DPNOM', 'DPMP']].itertuples(index=False, name=None))

# Combinaciones únicas en muertes
muertes_combinaciones = set(muertes[['Depto', 'MpNombre']].itertuples(index=False, name=None))
```

```python
solo_en_pob = pob_combinaciones - muertes_combinaciones
print(len(solo_en_pob))  # Para saber cuántas son
solo_en_pob  # Para verlas si quieres
```

```python
df_solo_en_pob = pd.DataFrame(list(solo_en_pob), columns=['DPNOM', 'DPMP'])
df_solo_en_pob = df_solo_en_pob.sort_values(by=['DPNOM', 'DPMP']).reset_index(drop=True)
df_solo_en_pob.to_excel('correcines-pob.xlsx')
df_solo_en_pob
```
