```python
bins_año = [1984, 1994, 2004, 2014]  
labels_año = ["1985-1994", "1995-2004", "2005-2014"]  
data["año_grupo"] = pd.cut(data["año_def"], bins=bins_año, labels=labels_año, right=True)

bins_edad = [0, 34, 54, 74, 100]  
labels_edad = ["Jóvenes (≤34)", "Mediana edad (35-54)", "Adultos mayores (55-74)", "Muy mayores (≥75)"]
data["edad_grupo"] = pd.cut(data["grupo_edad"].str.extract("(\d+)")[0].astype(int), bins=bins_edad, labels=labels_edad)

def calcular_frecuencia(variable, categoria, orden=None):
    tabla = data[variable].value_counts().reset_index()
    tabla.columns = ["Característica", "n"]
    tabla["%"] = (tabla["n"] / tabla["n"].sum()) * 100
    tabla["Categoría"] = categoria  

    if orden:
        cat_type = CategoricalDtype(categories=orden, ordered=True)
        tabla["Característica"] = tabla["Característica"].astype(cat_type)
        tabla = tabla.sort_values("Característica")  

    return tabla

orden_año = ["1985-1994", "1995-2004", "2005-2014"]
orden_edad = ["Jóvenes (≤34)", "Mediana edad (35-54)", "Adultos mayores (55-74)", "Muy mayores (≥75)"]
orden_seg_social = data["seg_social"].dropna().unique().tolist() 
orden_region = data["region"].dropna().unique().tolist()  

tabla_año = calcular_frecuencia("año_grupo", "Año de Defunción", orden_año)
tabla_edad = calcular_frecuencia("edad_grupo", "Edad", orden_edad)
tabla_seg_social = calcular_frecuencia("seg_social", "Seguridad Social", orden_seg_social)
tabla_region = calcular_frecuencia("region", "Región", orden_region)

tabla_final = pd.concat([tabla_año, tabla_edad, tabla_seg_social, tabla_region], ignore_index=True)

tabla_final = tabla_final[["Categoría", "Característica", "n", "%"]]

tabla_final
```
