```python
muertes_municipio = data.groupby(['Nombre_Departamento_Res', 'Nombre_Municipio_Res', 'año_def']).size()
muertes_municipio
```

```python
muertes_municipio_df = muertes_municipio.reset_index(name='conteo_muertes')
muertes_municipio_df
```

```python
gj_info = pd.read_excel('metadata_mun_col.xlsx')
gj_info.info()
```

```python
municipios_completos = gj_info[['MpNombre', 'Depto']].copy()
municipios_completos
```

```python
municipios_completos.query("Depto == 'VALLE DEL CAUCA' and  MpNombre == 'BUGA'")
```

```python

municipios_completos = gj_info[['MpNombre', 'Depto']].copy()

municipios_completos = municipios_completos.drop_duplicates()

años = pd.DataFrame({'Año': list(range(2009, 2024))})

municipios_expandidos = municipios_completos.merge(años, how='cross')

municipios_expandidos['Fallecidas'] = 0

municipios_expandidos

```

```python
muertes_municipio_df.query("Nombre_Departamento_Res == 'VALLE DEL CAUCA' and  Nombre_Municipio_Res == 'GUADALAJARA DE BUGA'")
```

```python
municipios_join = municipios_expandidos.merge(
    muertes_municipio_df,
    left_on=['Depto', 'MpNombre', 'Año'],
    right_on=['Nombre_Departamento_Res', 'Nombre_Municipio_Res', 'año_def'],
    how='left'
)
municipios_join
```

```python
muertes_mun_f = municipios_join.copy()
```

```python
muertes_mun_f['conteo_muertes'] = muertes_mun_f['conteo_muertes'].fillna(0)
```

```python
muertes_mun_f['Total_Fallecidas'] = muertes_mun_f['Fallecidas'] + muertes_mun_f['conteo_muertes']
```

```python
# Filtrar las filas con Total_Fallecidas > 0
muertes_mun_f_pos = muertes_mun_f.query('Total_Fallecidas > 0')

# Hacer merge para encontrar diferencias
diferencia = muertes_municipio_df.merge(
    muertes_mun_f_pos,
    left_on=['Nombre_Municipio_Res', 'Nombre_Departamento_Res', 'año_def'],
    right_on=['MpNombre', 'Depto', 'Anio'],
    how='left',
    indicator=True
)

# Filtrar las filas que están solo en muertes_municipio_df
solo_en_muertes_municipio_df = diferencia[diferencia['_merge'] == 'left_only']

# Resultado final sin columnas auxiliares del merge
solo_en_muertes_municipio_df = solo_en_muertes_municipio_df[muertes_municipio_df.columns]
solo_en_muertes_municipio_df
```
