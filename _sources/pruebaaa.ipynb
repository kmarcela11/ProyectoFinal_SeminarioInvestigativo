```python
import pandas as pd
from unidecode import unidecode
```

```python
data = pd.read_csv('resultados_finales.csv')
data['AÑO'] = '2009-2023'
dataq1 = pd.read_excel('tasas_quinquenio1.xlsx')
dataq2 = pd.read_excel('tasas_quinquenio2.xlsx')
dataq3 = pd.read_excel('tasas_quinquenio3.xlsx')
```

```python
dataq1
```

```python
data.rename(columns={'Nombre_Departamento_Def': 'DPNOM'}, inplace=True)
```

```python
dataq1['DPNOM'] = dataq1['DPNOM'].apply(unidecode)
reemplazo = {'ARCHIPIELAGO DE SAN ANDRES': 'SAN ANDRES PROVIDENCIA Y SANTA CATALINA', 'NARINO': 'NARIÑO'}
dataq1['DPNOM'] = dataq1['DPNOM'].replace(reemplazo)
dataq1['AÑO'] = '2009-2013'
dataq2['DPNOM'] = dataq2['DPNOM'].apply(unidecode)
dataq2['DPNOM'] = dataq2['DPNOM'].replace(reemplazo)
dataq2['AÑO'] = '2014-2018'
dataq3['DPNOM'] = dataq3['DPNOM'].apply(unidecode)
dataq3['DPNOM'] = dataq3['DPNOM'].replace(reemplazo)
dataq3['AÑO'] = '2019-2023'
```

```python
data_f = pd.concat([data,dataq1, dataq2, dataq3], ignore_index=True)
data_f
```

```python
dataq1.rename(columns={'Tasa_Mortalidad_Cruda': 'TMC Q1'}, inplace=True)
dataq1.rename(columns={'TAE': 'TAE Q1'}, inplace=True)
dataq1.rename(columns={'SMR': 'SMR Q1'}, inplace=True)
dataq1.rename(columns={'TOTAL_MUJERES': 'TOTAL_MUJERES Q1'}, inplace=True)
dataq1.rename(columns={'FALLECIDAS': 'FALLECIDAS Q1'}, inplace=True)
dataq1.rename(columns={'MUERTES_ESPERADAS': 'MUERTES_ESPERADAS Q1'}, inplace=True)

dataq2.rename(columns={'Tasa_Mortalidad_Cruda': 'TMC Q2'}, inplace=True)
dataq2.rename(columns={'TAE': 'TAE Q2'}, inplace=True)
dataq2.rename(columns={'SMR': 'SMR Q2'}, inplace=True)
dataq2.rename(columns={'TOTAL_MUJERES': 'TOTAL_MUJERES Q2'}, inplace=True)
dataq2.rename(columns={'FALLECIDAS': 'FALLECIDAS Q2'}, inplace=True)
dataq2.rename(columns={'MUERTES_ESPERADAS': 'MUERTES_ESPERADAS Q2'}, inplace=True)


dataq3.rename(columns={'Tasa_Mortalidad_Cruda': 'TMC Q3'}, inplace=True)
dataq3.rename(columns={'TAE': 'TAE Q3'}, inplace=True)
dataq3.rename(columns={'SMR': 'SMR Q3'}, inplace=True)
dataq3.rename(columns={'TOTAL_MUJERES': 'TOTAL_MUJERES Q3'}, inplace=True)
dataq3.rename(columns={'FALLECIDAS': 'FALLECIDAS Q3'}, inplace=True)
dataq3.rename(columns={'MUERTES_ESPERADAS': 'MUERTES_ESPERADAS Q3'}, inplace=True)

```

```python
data_final = pd.merge(data, dataq1, left_on='Nombre_Departamento_Def', right_on='DPNOM', how='left')
data_final.drop(columns=['DPNOM'], inplace=True)
data_final = pd.merge(data_final, dataq2, left_on='Nombre_Departamento_Def', right_on='DPNOM', how='left')
data_final.drop(columns=['DPNOM'], inplace=True)
data_final = pd.merge(data_final, dataq3, left_on='Nombre_Departamento_Def', right_on='DPNOM', how='left')
```

```python
data_final.to_excel('resultados_finales_con_tasas.xlsx', index=False)
```
