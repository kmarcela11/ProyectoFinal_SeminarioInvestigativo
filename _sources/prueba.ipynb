```python
import pandas as pd
```

```python
muertes = pd.read_csv('data_to_tasas.csv')
poblacion = pd.read_excel('POB-M40-MUNCOL-2009-2023.xlsx')
muertes = muertes.drop(columns=['Unnamed: 0'])

muertes = muertes.rename(columns={'Nombre_Departamento_Def': 'DPNOM', 'año_def': 'ANIO', 'Nombre_Municipio_Def': 'DPMP'}).copy()
equivalencias = {
    'SANTAFE DE BOGOTA D.C': 'BOGOTA DC',
    'BOGOTÁ, D.C.': 'BOGOTA DC',
    'ARCHIPIELAGO DE SAN ANDRES PROVIDENCIA Y SANTA CATALINA': 'ARCHIPIÉLAGO DE SAN ANDRÉS',
    'BOLIVAR': 'BOLÍVAR',
    'BOYACA': 'BOYACÁ',
    'ATLANTICO': 'ATLÁNTICO',
    'CORDOBA': 'CÓRDOBA',
    'GUAINIA': 'GUAINÍA',
    'GUAVIARE': 'GUAVIARE',
    'VAUPES': 'VAUPÉS',
    'VICHADA': 'VICHADA',
    'NORTE DE SANTANDER': 'NORTE DE SANTANDER',
    'CHOCÓ': 'CHOCÓ'
}

muertes['DPNOM'] = muertes['DPNOM'].replace(equivalencias)
poblacion['DPNOM'] = poblacion['DPNOM'].replace(equivalencias)
condicion = (muertes['DPNOM'] == 'BOGOTA DC') | (muertes['Nombre_Departamento_Res'] == 'BOGOTA DC')


```

```python
# Aplicar los cambios a las columnas DPNOM y Nombre_Departamento_Res para las filas que cumplen la condición
muertes.loc[condicion, ['DPNOM', 'Nombre_Departamento_Res']] = 'CUNDINAMARCA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'BUENAVISTA'), 'DPNOM'] = 'SUCRE'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'CALI'), 'DPNOM'] = 'VALLE DEL CAUCA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'CARTAGENA DE INDIAS'), 'DPNOM'] = 'BOLÍVAR'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'CHAPARRAL'), 'DPNOM'] = 'TOLIMA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'CHINAVITA'), 'DPNOM'] = 'BOYACÁ'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'CHIRIGUANÁ'), 'DPNOM'] = 'CESAR'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'DUITAMA'), 'DPNOM'] = 'BOYACÁ'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'ENVIGADO'), 'DPNOM'] = 'ANTIOQUIA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'GARZÓN'), 'DPNOM'] = 'HUILA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'IBAGUÉ'), 'DPNOM'] = 'TOLIMA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'MANIZALES'), 'DPNOM'] = 'CALDAS'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'MELGAR'), 'DPNOM'] = 'TOLIMA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'PEREIRA'), 'DPNOM'] = 'RISARALDA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'POPAYÁN'), 'DPNOM'] = 'CAUCA'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'PUERTO LÓPEZ'), 'DPNOM'] = 'META'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'TUNJA'), 'DPNOM'] = 'BOYACÁ'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'VALLEDUPAR'), 'DPNOM'] = 'CESAR'
muertes.loc[(muertes['DPNOM'] == 'CUNDINAMARCA') & (muertes['DPMP'] == 'VILLAVICENCIO'), 'DPNOM'] = 'META'
```

```python
# Obtener pares únicos
muertes_pares = muertes[['DPNOM', 'DPMP']].drop_duplicates()
poblacion_pares = poblacion[['DPNOM', 'DPMP']].drop_duplicates()

# Comparar: pares en muertes pero no en población
solo_muertes = muertes_pares.merge(poblacion_pares, on=['DPNOM', 'DPMP'], how='left', indicator=True)
solo_muertes = solo_muertes[solo_muertes['_merge'] == 'left_only']

# Comparar: pares en población pero no en muertes
solo_poblacion = poblacion_pares.merge(muertes_pares, on=['DPNOM', 'DPMP'], how='left', indicator=True)
solo_poblacion = solo_poblacion[solo_poblacion['_merge'] == 'left_only']
```

```python
muertes_set = set(tuple(x) for x in muertes[['DPNOM', 'DPMP']].drop_duplicates().values)
poblacion_set = set(tuple(x) for x in poblacion[['DPNOM', 'DPMP']].drop_duplicates().values)

# Diferencias
solo_en_muertes = muertes_set - poblacion_set
solo_en_poblacion = poblacion_set - muertes_set
```
